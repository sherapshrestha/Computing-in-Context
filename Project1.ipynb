```python
base_salaries = []

with open("2021_base_salary_police.csv", "r") as file:
    base_salary_police_2021 = csv.DictReader(file)

    for row in  base_salary_police_2021:
        base_salaries.append(float(row["Base Salary"]))
        
sorted_salaries = sorted(base_salaries)
n = len(sorted_salaries)

if n % 2 == 1:
    median_salary = sorted_salaries[n//2]
else:
    median_salary = (sorted_salaries[n // 2 -1] + sorted_salaries[n//2])/2

median_salary
```

the first code is to create an empty list to store the base salary values from each row of the CSV file.
the next two code is the same I have performed earlier. 
in order to find the median, sorting values is necessary. 
thats why the next code sorts the base salary list in an ascending order and it stores that value in sorted_salaries variable.
the next code will take the number of items in sorted_salaries and store it in n variable. 
the if function checks whether the variable 'n' which has number of items is odd, if it is the median would be the middle
item in the sorted_salaries var. 
else, if n is even, then the median value will be the average of two middle values.
lastly, the median_salary variable shows the value. 

```python
base_salaries = []

with open("2021_base_salary_police.csv", "r") as file:
    base_salary_police_2021 = csv.DictReader(file)
    
    for row in base_salary_police_2021:
        base_salaries.append(float(row["Base Salary"]))
        
salary_counts = {}

for salary in base_salaries:
    if salary in salary_counts:
        salary_counts[salary] += 1
    else:
        salary_counts[salary] = 1

mode_salary = 0
max_count = 0

for salary, count in salary_counts.items():
    if count > max_count:
        max_count = count
        mode_salary = salary

mode_salary
```

the first few lines of the code are repeated. 
salary_counts is a dictionary which will store each unique salary as a key and the no. of times it appears as its values.
for loop iterates over each salary in the base_salary list.
the if statemnt checks if salary is already a key in salary_counts, if it is, it increases its count by 1.
else, the code will add a new key with a count of 1.
mode_salary and max_count is initialized to 0. 
the next for loop iterates over each salary, count pair in salary_counts.
the if statement checks if the current count is greater than max count, if it is, max count is updated to count and mode salary will be set to the corresponding salary.
after this loop, the mode_salary variable will contain the salary that appears the most no. of times in the list. 
return the mode_salary variable to see its value.

```python
max_salary = df['Base Salary'].max()
```

max_salary variable is created accessing the base salary column in the dataframe and using a max method we find the maximum value in the base salary column. 

```python
def gen_sparkline(value, max_value, max_stars=10):
    stars_count = int(value / max_value * max_stars)
    return '*' * stars_count
```

we are defining a function gen sparkline which takes 3 parameters into account, the value parameter has the actual value to show.
max_value parameter is the max possible value which will help determind the scale.
max stars is the max no. of stars the function can display, i have chosen the default to be 10. 

```python
print("Fiscal Year    Sparkline")
for index, row in df.iterrows():
    sparkline = gen_sparkline(row['Base Salary'], max_salary)
    print(f"{row['Fiscal Year']}    {sparkline}")
```

for this code, the line prints the header for the output, i have used fiscal year and sparkline.
followed by a for loop, that iterates over the rows of the dataframe.
inside the loop, the sparkline generates a sparkline for each row based on the value in the base salary column and max salary is used as a parameter to compare. 
finally using f string to print the fiscal year from the row['fiscal year'] and its corresponding sparkline.
